### Like to Dislike Ratio

---
> Adding a new column to each dataset with the like-dislike ratio (likes divided by dislikes) for the videos in each dataset.
---

```python
us_df.head(2)
```

```python
us_df['like-dislike'] = us_df['likes']/ us_df['dislikes']
```

```python
us_df.head(3)
```

```python
gb_df['like-dislike'] = gb_df['likes'] / gb_df['dislikes']
```

```python
gb_df.head(3)
```

> Now looking into the average of the newly created column ratio like-dislike for all the videos in each dataset

```python
us_avg_ld = us_df[['like-dislike']].mean(skipna=True)
```

```python
us_avg_ld
```

```python
us_df['like-dislike'].shape
```

```python
us_df['like-dislike'].describe()
```

```python
us_df['like-dislike'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
us_df.dropna(inplace=True)
```

```python
us_df.shape
```

```python
us_df[['like-dislike']].mean()
```

> Here at the start we had some infinite values for the like to dislike ratio. We went ahaed and dropped them in order to have an actual value we can see and interpret. 
---
> In the end the I get a like-dislike average ratio for the US dataset to be 44

```python
gb_df[['like-dislike']].shape
```

```python
gb_df[['like-dislike']].describe()
```

```python
gb_df['like-dislike'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

## Business Report

> Client hypothesizes that the most polarizing videos get shared the most. Analyze the data to determine whether the client is correct, and what other metrics should be investigated. 

```python
us_df
```

Most polarizing videos would have to be ones with the lowest like-dislike ratio. This is a very important metric as well as looking into the views. We do not have a direct measurement of how often the video was shared. The only conclusion from this data is looking at the created ratio and views it obtains

```python
plt.scatter(us_df['like-dislike'], us_df['views'], c='green')
plt.xlabel('like-dislike')
plt.ylabel('views')
plt.title('Polarizing Videos US')
plt.show()
```

```python
plt.scatter(us_df['like-dislike'], us_df['views'], c='blue')
plt.xlabel('like-dislike')
plt.ylabel('views')
plt.title('Polarizing Videos GB')
plt.show()
```

> According to this graph with like-dislike vs. views the most views videos have the smallest like-dislike ratio. Meaning that the more polarizing (greater dislikes to likes) a videos receives the more views it has. With this there can be an assumption that it is more widely shared. Of course I would need more data with share count or unique viewers to have a concrete answer. 

> From these findings our client does have a strong point and data is able to back that point
